```python
# NLP library to perform stemming of 'tags' (Normalising similar words into the same entity)
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmerFn(text):
    new_tags = []
    for i in text.split():
        new_tags.append(ps.stem(i))
        
    return " ".join(new_tags)
```

```python
# Finding the average of the 'vote_average' values across all tuples
avgVote = movie_ratings['vote_average'].mean()
print(avgVote)
```

```python
# Finding the average of the 'vote_count' values across all tuples
avgVoteCount = movie_ratings['vote_count'].mean()
print(avgVoteCount)
```

```python
def voteBasedRating(movieList, avgVoteCount=avgVoteCount, avgVote=avgVote):
    voteCount = movieList['vote_count']
    voteRating = movieList['vote_average']
    # Weighted rating based on the IMDB formula
    return (voteCount/(voteCount+avgVoteCount) * voteRating) + (avgVoteCount/(voteCount+avgVoteCount) * avgVote)
```

```python
movie_ratings['score'] = movie_ratings.apply(voteBasedRating, axis=1)
```

```python
movie_ratings.head()
```

```python
# Sorting movies in descending order of 'score' & resetting the index values
movie_ratings = movie_ratings.sort_values(by=['score'], ascending=False)
movie_ratings = movie_ratings.reset_index(drop=True)
```

```python
movie_ratings.head(100)
```

```python
# Recommender function (vote based, filtered by genre)
def recommenderFnVoteBased(genre):
    # Removing spaces from genre & converting to lower case
    genre = genre.replace(" ","")
    genre = genre.lower()
    ct = 0
    recommendedList = []
    for movie in range(len(movie_ratings)):
        if movie_ratings.loc[movie,'genres'].find(genre) != -1:
            recommendedList.append({'title': movie_ratings.loc[movie,'original_title'], 'score': round(movie_ratings.loc[movie,'score'], 2)})
            ct += 1
            if ct == 3:
                break
    if(len(recommendedList) == 0):
        print('No movies found for the genre: ', genre)
    else:
        for movie in recommendedList:
            print(movie['title'], movie['score'])
    
    return
```
